```python
#Packages
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
#I/- DATA TREATMENT
data["FOOD"]=data["FOOD"].convert_objects(convert_numeric=True)
data["SERVICE"]=data["SERVICE"].convert_objects(convert_numeric=True)
data["VALUE"]=data["VALUE"].convert_objects(convert_numeric=True)
data["GRADE_MEAN_RESTAURANT"]=data["GRADE_MEAN_RESTAURANT"].convert_objects(convert_numeric=True)
data["GRADE"]=data["GRADE"].convert_objects(convert_numeric=True)
data["TYPES OF FOOD"]=data["TYPES OF FOOD"].convert_objects(convert_numeric=True)

# remplacing NaN in column Value, Service & Food with mean value
data["VALUE"]=(data["VALUE"]).fillna(data["GRADE_MEAN_VALUE"])
data["SERVICE"]=(data["SERVICE"]).fillna(data["GRADE_MEAN_SERVICE"])
data["FOOD"]=(data["FOOD"]).fillna(data["GRADE_MEAN_FOOD"])
data["GRADE"]=(data["GRADE"]).fillna(data["GRADE_MEAN_RESTAURANT"])
```

```python
#Extracting list of unique values for any column
def unique_column (data,column_name):
    array= []
    for x in data[column_name]:
        new=str.split(str(x),",")
        for i in new:
            if i not in array: 
                array.append(i)
    return(array)
```

```python
# adding new columns and binary values

def split_columns(data,column,names):
    for g in names:
        new_column=[]
        for i,x in enumerate(data[column]):
            new=str.split(str(x),",")
            new2=str(new).replace(" ", "")
            g2=str(g).replace(" ", "")
            if g2 in new2:
                new_column.append(1)
            else:
                new_column.append(0)

        data[str(g2)]= new_column
```

```python
#Comments treatment
comments=databis.copy()
from stop_words import get_stop_words
stop_en = get_stop_words('en')

comments["VALID"] = comments["COMMENT"].isnull()
comments=comments[~comments["VALID"]]
print(set(list(comments['COMMENT'].apply(lambda x: type(x)))))

def clean_text(x):
    words = x.lower().split()
    words = [mot for mot in words if not mot in stop_en]
    words = [mot for mot in words if len(mot)>1]
    return words

comments["CLEANED_TEXT"] = comments['COMMENT'].apply(clean_text)
```

```python
def conteur(x):
    c=0
    for mot in x:
        if mot=="great":
            c+=1
    return c

def conteur1(x):
    c=0
    for mot in x:
        if mot=="good":
            c+=1
    return c

def conteur2(x):
    c=0
    for mot in x:
        if mot=="nice":
            c+=1
    return c
def conteur3(x):
    c=0
    for mot in x:
        if mot=="delicious":
            c+=1
    return c

def conteur4(x):
    c=0
    for mot in x:
        if mot=="service":
            c+=1
    return c

def conteur5(x):
    c=0
    for mot in x:
        if mot=="wait":
            c+=1
    return c
```

```python
#Add of more features: Specific words 
comments['Great'] = comments['CLEANED_TEXT'].apply(lambda x: conteur(x))
comments['Good'] = comments['CLEANED_TEXT'].apply(lambda x: conteur1(x))
comments['Delicious'] = comments['CLEANED_TEXT'].apply(lambda x: conteur3(x))
comments['Nice'] = comments['CLEANED_TEXT'].apply(lambda x: conteur2(x))
comments['Service'] = comments['CLEANED_TEXT'].apply(lambda x: conteur4(x))
comments['Wait'] = comments['CLEANED_TEXT'].apply(lambda x: conteur5(x))
```

```python
columns = ['Unnamed: 0','Unnamed: 0.1','VALID','GOOD FOR','DINNING STYLE','FEATURES','Neighborhood']
comments.drop(columns, inplace=True, axis=1)
```
